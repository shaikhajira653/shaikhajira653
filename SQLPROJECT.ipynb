1.Print customer Id, customer name and average account_balance maintained by each customer for all of his/her accounts in the bank.(8 Rows)

```python
# SQL query to get customer_id, customer_name, and average balance amount
query = """
SELECT
    c.customer_id,
    c.customer_name,
    AVG(a.balance_amount) AS avg_balance
FROM BANK_CUSTOMER c
LEFT JOIN BANK_Account_Details a ON c.customer_id = a.customer_id
GROUP BY c.customer_id, c.customer_name
LIMIT 8;  -- Show only 8 rows
"""

# Execute the query
cursor.execute(query)

# Fetch column names
columns = [desc[0] for desc in cursor.description]

# Fetch all results
results = cursor.fetchall()

# Print table headers
print("{:<12} {:<20} {:<15}".format(*columns))
print("-" * 50)

# Print each row in a properly formatted way
for row in results:
    print("{:<12} {:<20} {:<15.2f}".format(*row))  # Format avg_balance to 2 decimal places
```

```python
#2) Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows).
query = """
SELECT
    a.customer_id,
    a.account_number,
    COALESCE(a.balance_amount, t.transaction_amount) AS balance_amount
FROM BANK_Account_Details a
LEFT JOIN BANK_Account_transaction t ON a.account_number = t.account_number
WHERE a.account_type = 'Credit Card'
LIMIT 4;  -- Show only 4 rows
"""
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers='keys', tablefmt='psql'))
```

```python
query = """
SELECT
    b.customer_id,
    b.account_number,
    b.account_type,
    b.balance_amount - IFNULL(SUM(CASE WHEN t.transaction_amount < 0 THEN ABS(t.transaction_amount) ELSE 0 END),0) AS deducted_balance
FROM
    Banked_Account_information b
LEFT JOIN
    banked_accounts_transaction t
ON
    b.account_number = t.account_number
WHERE
    b.relationship_type = 'P'
GROUP BY
    b.customer_id, b.account_number, b.account_type, b.balance_amount
ORDER BY
    b.customer_id;
"""

df = pd.read_sql_query(query, conn)

# 6. Display result
print(df.shape[0], "Rows returned")
df
```

```python
query = """
SELECT
    b.account_number AS savings_account,
    b.account_type AS account_type,
    SUM(t.transaction_amount) AS total_linked_transaction_amount
FROM Banked_Accounts_informations b
JOIN bank_transaction_relationship_details r
ON b.account_number = r.primary_account_number
JOIN Banked_Accounts_informations linked
ON r.linked_account_number = linked.account_number
JOIN banked_account_transactioned t
ON linked.account_number = t.account_number
WHERE b.account_type = 'SAVINGS'
AND linked.account_type = 'CREDIT CARD'
GROUP BY b.account_number, b.account_type;
"""

df = pd.read_sql_query(query, conn)

# 7. Display result
print(df.shape[0], "Rows returned")
df
```
