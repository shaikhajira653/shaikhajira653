```python
import re
import os

# File to store user data
USER_DATA_FILE = "user_data.txt"

# Function to validate username
def validate_username(username):
    pattern = r"^[A-Za-z][A-Za-z0-9._]*@[A-Za-z]+\.[A-Za-z.]+$"
    if re.match(pattern, username) and ".." not in username and "@." not in username:
        return True
    return False

# Function to validate password
def validate_password(password):
    if (6 <= len(password) <= 16 and
        re.search(r"[A-Z]", password) and
        re.search(r"[a-z]", password) and
        re.search(r"[0-9]", password) and
        re.search(r"[!@#$%^&*(),.?\":{}|<>]", password)):
        return True
    return False

# Function to store user data
def store_user_data(username, password):
    with open(USER_DATA_FILE, "a") as file:
        file.write(f"{username}:{password}\n")

# Function to get stored users
def get_stored_users():
    users = {}
    if os.path.exists(USER_DATA_FILE):
        with open(USER_DATA_FILE, "r") as file:
            for line in file:
                user, pwd = line.strip().split(":")
                users[user] = pwd
    return users

# Registration Process
def register():
    print("\n--- User Registration ---")
    while True:
        username = input("Enter a username (email format): ")
        if not validate_username(username):
            print("Invalid username format. Try again!")
            continue

        password = input("Enter a password: ")
        if not validate_password(password):
            print("Invalid password format! Must include uppercase, lowercase, number, special character, length 6-16.")
            continue

        users = get_stored_users()
        if username in users:
            print("Username already exists! Try logging in.")
            return
        store_user_data(username, password)
        print("Registration successful!\n")
        break

# Login Process
def login():
    print("\n--- User Login ---")
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    users = get_stored_users()
    if username in users and users[username] == password:
        print("Login successful! Welcome back.")
    else:
        print("Login failed! Invalid credentials.")
        forgot_or_register()

# Forgot Password or Register Option
def forgot_or_register():
    print("\nWhat would you like to do?")
    print("1. Register a new account")
    print("2. Forgot Password")
    choice = input("Enter 1 or 2: ")

    if choice == "1":
        register()
    elif choice == "2":
        forgot_password()
    else:
        print("Invalid choice.")

# Forgot Password Process
def forgot_password():
    print("\n--- Forgot Password ---")
    username = input("Enter your username: ")
    users = get_stored_users()

    if username in users:
        print(f"Your password is: {users[username]}")
        reset = input("Would you like to reset it? (yes/no): ").lower()
        if reset == "yes":
            new_password = input("Enter new password: ")
            if validate_password(new_password):
                users[username] = new_password
                with open(USER_DATA_FILE, "w") as file:
                    for user, pwd in users.items():
                        file.write(f"{user}:{pwd}\n")
                print("Password reset successful!")
            else:
                print("Password doesn't meet the criteria. Try again later.")
    else:
        print("Username not found. Please register.")
        register()

# Main Menu+++
def main():
    while True:
        print("\n--- Welcome to the Python Auth System ---")
        print("1. Register")
        print("2. Login")
        print("3. Exit")
        choice = input("Enter your choice (1/2/3): ")

        if choice == "1":
            register()
        elif choice == "2":
            login()
        elif choice == "3":
            print("Exiting system... Goodbye!")
            break
        else:
            print("Invalid choice! Try again.")

# Run the program
if __name__ == "__main__":
    main()
```
